```python
#Quarta questão

import math

# Dados do problema
altura_total = 40  # cm
diametro_superior = 40  # cm
diametro_inferior = 30  # cm
altura_acumulada = 10  # cm
duracao_minutos = 110  # minutos

# Cálculo dos raios
raio_superior = diametro_superior / 2  # cm
raio_inferior = diametro_inferior / 2  # cm

# Volume total do tronco de cone
volume_total = (math.pi * altura_total / 3) * (raio_superior**2 + raio_superior * raio_inferior + raio_inferior**2)

# Volume acumulado (proporcional à altura acumulada)
volume_acumulado = (altura_acumulada / altura_total) * volume_total

# Área superior do balde
area_superior = math.pi * raio_superior**2

# Duração em horas
duracao_horas = duracao_minutos / 60

# Precipitação média (mm/h)
# Volume acumulado é em cm³, precisamos converter para mm³ e dividir pela área em mm²
volume_acumulado_mm3 = volume_acumulado * 1000  # Conversão de cm³ para mm³
area_superior_mm2 = area_superior * 100  # Conversão de cm² para mm²
precipitacao_media = volume_acumulado_mm3 / (area_superior_mm2 * duracao_horas)

# Resultado
print(f"Volume total do tronco de cone: {volume_total:.2f} cm³")
print(f"Volume acumulado (10 cm): {volume_acumulado:.2f} cm³")
print(f"Área superior do balde: {area_superior:.2f} cm²")
print(f"Taxa média de precipitação: {precipitacao_media:.2f} mm/h")
```
