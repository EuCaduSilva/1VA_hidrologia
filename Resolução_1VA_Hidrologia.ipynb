```python
#Resolução primeira questão
#Precipitação média Anual(P)(mm):
P= 1400

#Vazão média do rio(Q)(m³s-1)
Q= 40

#Área da bacia(A)(m²)
A = (3.5*10**9)

#Lamina escoada(Le)(mm)
Le= (((Q*3600*24*365)/A)*1000)

#Evapotranspiração média anual
Ev = P - Le

#resultados
print('\n','P=',P,'mm','\n','Q=',Q,'m³s-1','\n','A=',A,'m²','\n','Le=', Le,'mm')
print('1.1 Evapotranspiração anual=', Ev,'mm')

#coeficiente de escoamento(C)
C=(Le/P)*100
print('1.2 Coeficiente de escoamento=',C,'%')
```
